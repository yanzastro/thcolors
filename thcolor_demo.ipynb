```python
from thcolor import *
import matplotlib.pyplot as plt
```

```python
def get_names():
    '''
    Get the name list of available characters.
    '''
    name_list = ['hina']
    for name_,dict_ in c.items():
        name = name_[:-1]
        if name == name_list[-1]:
            continue
        name_list.append(name)
    return name_list
```

```python
names = get_names()
```

```python
for name in names:
    fig, ax = plt.subplots(ncols=1, nrows=1)
    plot_demo(name, ax)
```
